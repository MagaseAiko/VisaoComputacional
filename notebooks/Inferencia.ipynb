```python
import os
from ultralytics import YOLO

import shutil
from IPython.display import Image, display
import glob
import cv2
```

```python
if os.path.exists('modelo_banana.pt'):
    print("Modelo copiado com sucesso! ✅")
else:
    print("Arquivo não encontrado. Verifique o caminho. ❌")
```

```python
nome_modelo = 'dataset_banana.pt'

print(f"Carregando modelo: {nome_modelo}...")
model = YOLO(nome_modelo)

print("\nClasses que este modelo conhece:", model.names)
```

```python
source = 'banana.mp4'

if not os.path.exists(source) and type(source) is str:
    print("Aviso: Vídeo local não encontrado, testando com imagem de bananas online...")
    source = 'https://upload.wikimedia.org/wikipedia/commons/thumb/a/af/Bananas_%28Alabama_Extension%29.jpg/330px-Bananas_%28Alabama_Extension%29.jpg'


results = model.predict(
    source=source,
    save=True,    
    conf=0.5,
    imgsz=320,   
    stream=True,   
    vid_stride=2    
)

for _ in results:
    pass

print("Processamento concluído!")
```

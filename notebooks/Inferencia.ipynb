```python
# Importações necessárias
import os
from ultralytics import YOLO
import glob
```

```python
# Nome do modelo treinado
if os.path.exists('model_banana.pt'):
    print("Modelo copiado com sucesso! ✅")
else:
    print("Arquivo não encontrado. Verifique o caminho. ❌")
```

```python
# Carregando o modelo
nome_modelo = 'model_banana.pt'

print(f"Carregando modelo: {nome_modelo}...")
model = YOLO(nome_modelo)

print("\nClasses que este modelo conhece:", model.names)
```

```python
# Realizar a inferência
source = 'banana.mp4'

if not os.path.exists(source) and type(source) is str:
    print("Aviso: Vídeo local não encontrado, testando com imagem de bananas online...")
    source = 'https://upload.wikimedia.org/wikipedia/commons/thumb/a/af/Bananas_%28Alabama_Extension%29.jpg/330px-Bananas_%28Alabama_Extension%29.jpg'

results = model.predict(
    source=source,
    save=True,    
    conf=0.55,
    imgsz=320,   
    stream=True,   
    vid_stride=2    
)

for _ in results:
    pass

print("Processamento concluído!")
```

```python
# Converter AVI para MP4 e remover o arquivo AVI
videos = glob.glob("runs/detect/predict*/*.avi")

if videos:
    input_video = videos[-1]
    output_video = input_video.replace(".avi", ".mp4")

    os.system(f'ffmpeg -i "{input_video}" -vcodec libx264 "{output_video}"')

    if os.path.exists(output_video):
        os.remove(input_video)
        print("MP4 salvo em:", output_video)
        print("AVI removido com sucesso.")
    else:
        print("Erro na conversão. AVI mantido.")
else:
    print("Nenhum vídeo AVI encontrado.")
```

```python
from ultralytics import YOLO

# Em vez de 'yolov8n.pt', carregamos O NOSSO ARQUIVO!
nome_modelo = 'modelo_aquario.pt'

print(f"Carregando modelo: {nome_modelo}...")
model = YOLO(nome_modelo)

# Verificar quais classes esse modelo conhece
print("\nClasses que este modelo conhece:", model.names)
```

```python
import cv2

# Defina um arquivo de origem de teste. Ex: um vídeo baixado do youtube
source = 'aquario.mp4'

# Se o arquivo local não existir, usa uma URL de teste com peixes
if not os.path.exists(source) and type(source) is str:
    print("Aviso: Vídeo local não encontrado, testando com imagem de peixes online...")
    source = 'https://upload.wikimedia.org/wikipedia/commons/e/e4/A_Shoal_of_Fish.jpg'

# Realiza a inferência e exibe (se rodando local em um ambiente com interface gráfica)
# conf=0.5 significa que só mostra se tiver 50% de certeza
results = model.predict(source=source, show=True, conf=0.5)
```
